````python
# NAN 변수 제거
elect_df = elect_df.iloc[:,1:]
elect_df
````

## 4. <b>다음을 만족하는 20년 달력 데이터 프레임을 만드시오.</b>

```
original    Year    Month    Day    weekday    weekend    season    
0   20-01-01     20       1       1       sat         1       winter
1   20-01-02     20       1       2       sun         1       winter
2   20-01-03     20       1       3       mon         0       winter

...   ...        ...     ...     ...      ...        ...        ...   

364 20-12-31     20       12      31      thr         0       winter

```

````python
import datetime
````

````python
date_df = pd.DataFrame()
# 2020년 1월 1일 ~ 2020년 12월 31일까지의 값 original변수에 넣기
original = pd.date_range(start='2020-01-01', end='2020-12-31')
date_df['original'] = original
date_df['Weekend'] = 0
date_df['Season'] = ''
````

````python
# Year, Month, Day 
date_df['Year'] =  date_df['original'].dt.year
date_df['Month'] = date_df['original'].dt.month
date_df['Day'] = date_df['original'].dt.day
date_df['Weekday'] = date_df['original'].dt.day_name()
````

````python
# 주말:1 평일:0
for i in range(len(date_df)):
    if date_df['Weekday'][i] == 'Saturday' or date_df['Weekday'][i] == 'Sunday':
        date_df['Weekend'][i] = 1
    else:
        date_df['Weekend'][i] = 0
````

````python
# 봄: 3,4,5 여름: 6,7,8 가을: 9,10,11 겨울: 12,1,2
for i in range(len(date_df)):
    if date_df['Month'][i] == 3 or date_df['Month'][i] == 4 or date_df['Month'][i] == 5:
        date_df['Season'][i] = 'Spring'
    elif date_df['Month'][i] == 6 or date_df['Month'][i] == 7 or date_df['Month'][i] == 8:
        date_df['Season'][i] = 'Summer'
    elif date_df['Month'][i] == 9 or date_df['Month'][i] == 10 or date_df['Month'][i] == 11:
        date_df['Season'][i] = 'Fall'
    else:
        date_df['Season'][i] = 'Winter'
````

````python
date_df = date_df[['original', 'Year', 'Month', 'Day', 'Weekday', 'Weekend', 'Season']]
date_df
````
